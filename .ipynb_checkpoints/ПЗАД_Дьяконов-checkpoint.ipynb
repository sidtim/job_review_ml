# Искусство визуализации (часть 2 - одномерный анализ). 
# [Видео №1](https://www.youtube.com/watch?v=kOaMvRo2YPI&list=PLaRUeIuewv8CMFox0oEjlyePUhUmo-x0h&index=6)
# [Видео №2. Продолжение.](https://www.youtube.com/watch?v=euK2xBgLHIg&list=PLaRUeIuewv8CMFox0oEjlyePUhUmo-x0h&index=7)
Одномерный анализ (Univariate Analysis) - исследование отдельных признаков (здесь: одномерные графики).
1. В одномерном анализе необходимо:
* Установить природу признаков. Проверить логичность признаков.
* У каждого признака есть ряд характеристик, которые нужно определить:
    * имя
    * область значений
    * распределение
    * особенности (аномалии, пропуски и т.п.)
    * устойчивость - характеристика, которая показывает то, насколько распределение признака может смещаться относительно времени и т.д. (например доходы клиентов банка)
    * важность - характеристика, которая показывает насколько признак хорош в какой-то конкретной задаче машинного обучения
* Визуализация признаков:
    * статистики признаков (описательные из математической статистики)
    * характеристики признаков (важности, AUC и т.п.)
    
***Важно: для описательных статистик нам необходим вариационный ряд***
    
2. Описательные статистики - среднее:
    * Выборочное среднее
    * Усеченное среднее
    * Медиана
    * Мода
    * Mid-range
    * Midhinge
3. Описательные статистики - характерные элементы:
    * Минимум / Максимум
    * Квантиль - значение, которое заданная с.в. не превышает с заданной вероятностью. Существуют разные квантили, а именно: кваритил, децили, процентили, n-ый элемент ($nth(X,k) = x_k$)
4. Описательные статистики - разброс значений (13:50):
    * Среднее линейное (абсолютное) отклонение. Mean Absolute Deviation.
    * Среднеквадратическое отклонение. MSE / MSD (Mean Squared Error / Mean Squared Deviation).
5. Описательные статистики - абсолютные вариации:
    * Стандартное отклонение: $std$
    * Размах: $range(X)=x_m-x_1$
    * Дисперсия: $std^2$
    * Median Absolute Deviation (MAD)
    * Среднее квартильное расстояние (Интерквартильный размах IQR)
6. Описательные статистики - относительные вариации (абсолютная вариация / среднее):
    * Коэффициент вариации(coefficient of variation): $\frac{std(X)}{mean(X)}$
    * Индекс дисперсии (Index of dispersion): $\frac{std^2(X)}{mean(X)}$
    * Относительный размах вариации (коэффициент осцилляции): $\frac{range(X)}{mean(X)}$
7. Описательные статистики - центральные моменты. (18:30)
8. Описательные статистики - моменты. (18:30)
9. Описательные статистики - стандартизованные моменты. (19:45)
10. Классный инсайт про бинарные признаки: в случае, если признак бинарный, то точка на графике в осях его среднего значения и стандартного отклонения будет располагаться на дугообразной кривой. Это происходит, потому что для бинарного признака $mean=p$, $std=\sqrt{p(1-p)}$ (30:10)
11. Что надо сразу выяснить про признак:
    * распределение значений признака
    * распределение обучение / тест
    * распределение целевой переменной (ex: класс 0 / 1)
    * **такие же шаги необходимо проделать для пропусков / выбросов**
12. Визуализация отдельных признаков:
    * Для визуализации отдельных признаков лучше строить гистограммы (причём гистограммы предпочтительней плотностей)
    * При построении гистограмм следует менять некоторые значения по умолчанию, а именно:
        * Число бинов (методы по установке числа бинов см.12:50)
            * Самый распространенный метод - подбирать число бинов. Также можно накладывать графики с разным числом бинов друг на друга, чобы увидеть отличия (13:25)
            * Число бинов = число уникльных значений признака
            * Число бинов = max(x) - min(x) ***Не совсем понял, почему такое значение оптимально***
        * Ширина столбцов

13. Анализ распределения можно осуществлять с помощью Q-Q plot.
14. Визуализация категориальных признаков:
    * Мелкие категории (которых крайне мало в выборке) можно перевести в категорию "Остальное".
    * Если информации для визуализации мало (т.е. мало категорий), то рекомендуется просто нарисовать таблицу с распределением категорий в выборке.
    * Можно логарифмировать гистограмму распределений категориальных признаков (30:10). Это полезно в случае, если у нас в категориальном признаке ОЧЕНЬ МНОГО ГРУПП (30.000 например). При логарифмировании мы сделаем график более плавным и более интерпретируемым.
